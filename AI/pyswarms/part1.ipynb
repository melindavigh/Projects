```python
import matplotlib.pyplot as plt
import numpy as np
import pyswarms as ps
```

```python

#Question 1
# Given a function y = -(x + sin(x))*exp(-x2), make a simple plot of this function.
# To plot this function, you need two arrays, X, Y. 
# X is already given as follows,
# X = np.linspace(-10, 10, 100)
# Use np.sin() for sin function and np.exp() for exponential.
# Be sure to include a title for this plot and labels for both x and y axis. The title could simply be “a plot for the test function’. 

# Define the function
def test_function(x):
    return -(x + np.sin(x)) * np.exp(-x**2)

# Define the x values
X = np.linspace(-10, 10, 100)

# Calculate the y values
Y = test_function(X)

# Create the plot
plt.plot(X, Y)
plt.grid(True)
plt.title("A plot for the test function")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Question 2
# Given the bounds is [-2.5, 2.5], use PySwarms to find the minimum of this function. 
# Be sure to enforce the right dimensionality for both the bounds and the settings of the Swarm algorithm. 
# Use the default ‘options’ for this problem and set the number of iterations to 1000. 

# Define function
def test_function(x):
    return -np.sum((x + np.sin(x)) * np.exp(-x**2))

# Set up hyperparameters
options = {'c1': 0.5, 'c2': 0.3, 'w': 0.9}

# Bounds
max_bound = 2.5 * np.ones(2) # return a tuple
min_bound = - 1 * max_bound
bounds = (min_bound, max_bound)

# Call instance of PSO
optimizer = ps.single.GlobalBestPSO(n_particles=20, dimensions=2, options=options, bounds=bounds)
# Perform optimization
cost, pos = optimizer.optimize(test_function, iters=1000)

# Print results
print("Minimum cost: ", cost)
print("Optimal position: ", pos)
```
